# Calculating VIF

### Two Techniques: 1. Collinearity check & 2. Residual Analysis

#### 1.Collinearity Check 

```python
rsq_r=smf.ols("RDS~ADMN+MS", data=Dataset).fit().rsquared
vif_r=1/(1-rsq_r)

rsq_a=smf.ols("ADMN~RDS+MS", data=Dataset).fit().rsquared
vif_a=1/(1-rsq_a)

rsq_m=smf.ols("MS~RDS+ADMN", data=Dataset).fit().rsquared
vif_m=1/(1-rsq_m)

# Storing vif values in a data frame
d1={'Varaibles':['RDS','ADMN','MS'],'Vif':[vif_r, vif_a, vif_m]}
Vif_DF=pd.DataFrame(d1)
Vif_DF
```

#### 2.Residual Analysis

```python
# Test for normality (Q-Q Plot)

sm.qqplot(model.resid,line='q')
plt.title("Normal Q-Q plot of residuals")
plt.show()
```

### Residual Plot for Homoscedasticity

```python
def get_standardized_values( vals ):
    return (vals - vals.mean())/vals.std()
```

```python
plt.scatter(get_standardized_values(model.fittedvalues),
            get_standardized_values(model.resid))

plt.title('Residual Plot')
plt.xlabel('Standardized Fitted values')
plt.ylabel('Standardized residual values')
plt.show()
```
